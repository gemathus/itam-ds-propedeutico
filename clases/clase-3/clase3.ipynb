#### Numpy

```python
import matplotlib.pyplot as plt
from numpy import arange,sin,cos
```

```python
x = arange(0.0,6.2,0.2)
x
```

```python
plt.plot(x,sin(x),'o-',x,cos(x),'^-')
plt.xlabel('x')
plt.legend(('sine','cosine'),loc = 0)
plt.grid(True)
plt.show()
```
