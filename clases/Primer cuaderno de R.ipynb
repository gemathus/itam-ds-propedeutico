```r
lanza_moneda <- function(numero_de_lanzamientos) {
    set.seed(2000)
    sample(c('aguila', 'sol'), numero_de_lanzamientos, replace=T)
}
```

```r
lanza_moneda(4)
```

```r
num_lanzamientos = 1000
sum(lanza_moneda(num_lanzamientos) == 'sol')/num_lanzamientos
```
